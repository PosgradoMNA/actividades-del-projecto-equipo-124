## ***Metodología CRISP-DM***
(Cross Industry Standard Process for Data Minig)

###**Se describe en 6 fases:**




###**1.   Entendimiento del Negocio**
En esta primera fase se establecen claramente los objetivos del proyecto, el tipo de análisis, qué alcance tendrá el proyecto y cuales serán los entregables. Todo esto desde el enfoque del cliente. 






###**2.   Entendimiento de los Datos**
En esta fase se lleva a cabo la recolección inicial de los datos. Se procede a explorar los datos, identificar variables clave, buscar errores o datos faltantes. Se incluyen también revisiones de la calidad de los datos.





###**3.   Preparación de los Datos**
En esta fase se consolidará el conjunto de datos final. Se realiza la limpieza de datos, reemplazar variables faltantes o remover duplicados y se le da el formato requerido. Comunmente esta es la fase mas extensa. 





###**4.   Modelado**
A continuación se seleccionan los modelos apropiados, por ejemplo, regresión lineal, arboles de decision, bosque aleatorio, etc. Posteriormente se construye y evalua el mejor modelo. De ser necesario, se puede dar una iteración entre ésta y la fase anterior.





###**5.   Evaluación**
En esta fase se evalúa de manera mas general los resultados del modelo y el proceso. Se revisa minuciosamente el modelo, se determina si los resultados son los esperados y si cumple con los objetivos. Posteriormente, se definen los pasos a seguir.





###**6.   Despliegue**
En esta última fase se despliega el modelo y se genera un plan de monitoreo y mantenimiento. Se produce un reporte final en donde se documentan todos los pasos realizados durante la preparación de los datos y modelado. Finalmente se hace una revisión general del proyecto.
